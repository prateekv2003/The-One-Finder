```python
regex_target = []
def regex(doc):
    reg = re.compile(r"\d{1,9}")
    for i in doc:
#         print(i)
        new = re.sub(reg, '', i)
        new = list(set(new.split()))
        regex_target.append(new)
    return regex_target
```
